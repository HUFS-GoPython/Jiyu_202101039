# 반복문의 종류 
- while 
- for 

## while 
- 어떤 조건이 만족하는 동안 명령 블록을 수행한다. 
- 조건문 + 반복문 (if + for loop)
- 해당 조건이 거짓이면 멈춘다. 
- 반복이 언제 멈추는지 그 시점을 정확히 파악해야 한다. 

## for vs while 
- for 
    - 반복 횟수를 미리 정한다. 
    - 그래서 정확하고 변하지 않는다. 
    - ex) 성적 산출 
- while 
    - 반복 횟수를 정하지 않는다. 
    - 반복 횟수가 중요하지 않다. 
    - 어떤 조건을 만족시키는 경우 반복을 중단할 때 사용한다. 
    - ex) 가위바위보

```python
# for loop

for i in range(3):
    print(i)
```

```python
#for i in range(3):
    #print(i) 를 while loop으로 

i = 0
while i <3:
    print(i)
    i += 1
```

```python
#for loop

abcd = ['a', 'b', 'c', 'd']
for ch in abcd:
    print(ch)
```

```python
#while loop

i = 0
while i < len(abcd):
    print(abcd[i])
    i += 1 
```

## 반복문 제어
- 되도록 사용하지 않는다. 
- 의도치 않은 오류를 발생시킬 수 있다. 

- break 
    - 논리적으로 반복을 종료시킬 때 사용한다. 
- else 
    - 어떤 조건이 완전히 끝났을 때 한 번 더 실행한다. 
    - break가 사용되었을 때에는 실행되지 않는다. 
    - 어느 지점에서 종료되었는지 확인할 때 사용한다. 
    - for & while 과 같은 들여쓰기 라인에서 사용한다. 
- continue 
    - 특정 조건을 건너뛰고 다음 반복을 실행할 때 사용한다. 

```python
i = 0
while i < 10:
    print(i)
    if i ==5 : break #반복문을 종료/ while문을 빠져나온다. 
    i += 1
```

```python
for i in range(10):
    if i == 5: break 
    print(i) #break와 print의 순서에 따라 결과가 달라짐
```

```python
i = 0
while i < 5:
    print(i)
    if i ==3: break
    i += 1
else: #break가 사용되면 else를 실행하지 않는다. 
    print('ended')
```

```python
for i in range(3):
    if i == 5: break 
    print(i)
else:
    print('ended') #for loop이 완전히 종료되었을 때 실행됨 
```

```python
for i in range(5):
    if i == 3: continue #3은 출력되지 않는다./ 특정 조건문을 건너 뛰고 다음 반복을 실행 
    print(i)
else:
    print('ended')
```

## 구구단 Quiz
- 변수 n, m 은 입력을 받아서 사용한다. 
- 메뉴 
    1. n단 출력 
    2. 1단부터 n단 출력 
    3. n단부터 m단 출력 

```python
while True:
    print("""메뉴\n 1.n단 출력 \n 2.1단부터 n단까지 \n 3, n단부터 m단까지 \n 4.종료""")
    menu = int(input(('구구단 메뉴를 선택하세요:')))
    if menu == 1:
        n = int(input('몇단을 출력하시겠습니까?:'))
        if n < 1 or n>9:
            print('잘못된 입력입니다.')
            continue
        print(f'{n}단을 출력하겠습니다.')
        for i in range(1,10):
            print(n,'X',i,'=',n*i)
    elif menu == 2:
        n = int(input('몇단까지 출력하시겠습니까?:'))
        print(f'{n}단까지 출력하겠습니다.')
        if n < 1 or n>9:
            print('잘못된 입력입니다.')
            continue
        for i in range(1,n+1):
            print(f'====={i}단======')
            for j in range(1,10):
                print(i,'X', j, '=', i*j)
    elif menu == 3:
        n = int(input('몇단부터 출력하시겠습니까?'))
        m = int(input('몇단까지 출력하시겠습니까?'))
        print(f'{n}단 부터 {m}단까지 출력하겠습니다.')
        if n > m:
            print('잘못된 입력입니다.')
            continue
        if n < 1 or n>9:
            print('잘못된 입력입니다.')
            continue
        if m <1 or m >9:
            print('잘못된 입력입니다.')
            continue
        for i in range(n,m+1):
            print(f'======{i}단=======')
            for j in range(10):
                print(i,'X', j, '=', i*j)
    elif menu == 4:
        print('종료')
        break
    else: 
        print('입력하신 번호는 메뉴에 없습니다.')
        continue
```

## Quiz 2
- 각 파일마다 word, character, sents 개수 출력하기

```python
for file in gutenberg.fileids():
    words = gutenberg.words(file)
    characters = gutenberg.raw(file)
    sents = gutenberg.sents(file)
    
    print(f'title:{file}\n words:{len(words)}, character:{len(characters)}, sentences:{len(sents)} ')
```

## Category - Frequency Distribution
- 'adventure' 카테고리의 frequency distribution을 구한다.

```python
from nltk import FreqDist
fdist = FreqDist([w.lower() for w in brown.words(categories ='adventure')])
fdist 
```

```python
## modal verbs
modals = ['can','could','will','would']
for verb in modals:
    print(verb, fdist[verb])
```
